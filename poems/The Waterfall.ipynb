__The Waterfall__

_for May Swenson_

For all they said,

&nbsp;&nbsp;I could not see the waterfall

&nbsp;&nbsp;&nbsp;&nbsp;until I came and saw the water falling,

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;its lace legs and its womanly arms sheeting down,


&nbsp;


while something howled like thunder,

&nbsp;&nbsp;over the rocks,

&nbsp;&nbsp;&nbsp;&nbsp;all day and all night -

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;unspooling

&nbsp;

like ribbons made of snow,

&nbsp;&nbsp;or god's white hair.

&nbsp;&nbsp;&nbsp;&nbsp;At any distance

it fell without a break or seam, and slowly, a simple

&nbsp;

preponderance––

&nbsp;&nbsp;a fall of flowers––and truly it seemed

&nbsp;&nbsp;&nbsp;&nbsp;surprised by the unexpected kindness of the air and

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;light-hearted to be

&nbsp;&nbsp;

flying at last.

&nbsp;&nbsp;Gravity is a fact everybody

&nbsp;&nbsp;&nbsp;&nbsp;knows about.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;It is always underfoot,


&nbsp;&nbsp;


like a summons,

&nbsp;&nbsp;gravel-backed and mossy,

&nbsp;&nbsp;&nbsp;&nbsp;in every beetled basin -

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;and imagination -


&nbsp;&nbsp;

that striver,

&nbsp;&nbsp;that third eye––

&nbsp;&nbsp;&nbsp;&nbsp;can do a lot but

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;hardly everything. The white, scrolled

&nbsp;&nbsp;

wings of the tumbling water

&nbsp;&nbsp;I never could have

&nbsp;&nbsp;&nbsp;&nbsp;imagined. And maybe there will be,

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;after all,

&nbsp;&nbsp;

some slack and perfectly balanced

&nbsp;&nbsp;blind and rough peace, finally,

&nbsp;&nbsp;&nbsp;&nbsp;in the deep and green and utterly motionless pools after all that

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;falling?

&nbsp;&nbsp;

Mary Oliver
